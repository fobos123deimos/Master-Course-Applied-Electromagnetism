```python
#@title b) $f_{0} = 10$ GHz

f_min = 5 #@param
f_max = 30 #@param
n_size = 1000 #@param
f0_value = 10 #@param
e_r2 = 3.7 #@param

frequencies = np.linspace(f_min, f_max, n_size)  # Frequencies from 5 GHz to 30 GHz
Gamma12_value =  (1-np.sqrt(e_r2))/(1+np.sqrt(e_r2)) # Value of Gamma12
Gamma23_value = -Gamma12_value  # Value of Gamma23



# Calculate |Î_in| for each frequency and convert to float
gamma_in_values = []
for freq in frequencies:
    value = abs((Gamma12_value + Gamma23_value * np.exp(-1j * 2*np.pi * freq / f0_value)) / (1 + Gamma12_value * Gamma23_value * np.exp(-1j * 2 * np.pi * freq / f0_value)))
    gamma_in_values.append(value)


# Create the plot with Plotly
fig = px.line(x=frequencies, y=gamma_in_values, labels={'x': 'Frequency (GHz)', 'y': 'Magnitude of the Input reflection coefficient |Î_in|'})
fig.update_layout(title='Magnitude of the Input Reflection Coefficient vs. Frequency', xaxis_title='Frequency (GHz)', yaxis_title='Magnitude of the Input reflection coefficient |Î_in|')
fig.show()
```

### Exemple 5-11 : plots for two sections

$$$$
### Binomial
$$$$

1. $$\Gamma_{in}(f) \simeq \displaystyle\sum^{N}_{n=0}\Gamma_{n} e^{-j2n\theta} = e^{-jN\theta}\Bigg(\frac{\eta_{L} - \eta_{0}}{\eta_{L}+\eta_{0}}\Bigg)cos^{N}(\theta) \simeq 2^{-N}\Bigg(\frac{\eta_{L} - \eta_{0}}{\eta_{L}+\eta_{0}}\Bigg)\sum^{N}_{n=0}C^{N}_{n}e^{-j2n\theta}$$

$$$$

2. $$C^{N}_{n} = \displaystyle\frac{N!}{(N-n)!n!} \qquad n = 0,1,2,...,N$$

$$$$

3. $$\Gamma_{n} = 2^{-N}\Bigg(\frac{\eta_{L} - \eta_{0}}{\eta_{L}+\eta_{0}}\Bigg)C^{N}_{n}$$

$$$$

4. $$\theta = \beta_{n}d_{n} = \displaystyle \Bigg(\frac{2\pi}{\lambda_{n}}\Bigg)\Bigg(\frac{\lambda_{n_{0}}}{4}\Bigg)=\displaystyle\frac{\pi}{2}\Bigg(\frac{f}{f_{0}}\Bigg)$$

$$$$

5. $$5 GHz \le f \le 30GHz$$

$$$$

6. $$Ïµ_{r_{3}} = Ïµ_{r_{2}} + 3 = 6.7$$

```python
#@title a) $f_{0} = 10$ GHz

f_min = 5 #@param
f_max = 30 #@param
n_size = 1000 #@param
f0_value = 10 #@param
e_L = 6.7 #@param
N = 2 #@param

frequencies = np.linspace(f_min, f_max, n_size)
eta_L = 377/np.sqrt(e_L)

exp_1 = (eta_L-377)/(eta_L+377)

C_n = lambda n: (math.factorial(N)/(math.factorial(N-n)*math.factorial(n)))

bin_aproximation = lambda fm : 2**(-N) * exp_1 * sum([ C_n(n)*np.exp(-1j*n*np.pi * fm / f0_value) for n in range(N+1) ])


# Calculate |Î_in| for each frequency and convert to float
gamma_in_values = []
for freq in frequencies:
    value = abs(bin_aproximation(freq))
    gamma_in_values.append(value)

relative_frequencies = frequencies/f0_value

# Create the plot with Plotly
fig = px.line(x=relative_frequencies, y=gamma_in_values, labels={'x': 'Relative Frequency (f/f0)', 'y': 'Magnitude of the Input reflection coefficient |Î_in|'})
fig.update_layout(title='Magnitude of the Input Reflection Coefficient vs. Relative Frequency (f/f0)', xaxis_title='Relative Frequency (f/f0)', yaxis_title='Magnitude of the Input reflection coefficient |Î_in|')
fig.show()
```

### Exemple 5-11 : plots for two sections

$$$$
### Tschebyscheff
$$$$

1. $$\Gamma_{in}(f) = e^{-jN\theta}\Bigg\{\frac{\eta_{L} - \eta_{0}}{\eta_{L}+\eta_{0}}\Bigg\}\Bigg\{ \frac{T_{N}[sec(\theta_{m})cos(\theta)]}{T_{N}[sec(\theta_{m})]}\Bigg\}$$

$$$$

2. $$T_{m}(z) = cos[m*arcos(z)] \qquad z \in [-1,+1] \quad ; \qquad T_{m}(z) = cosh[m*arcosh(z)] \qquad z>+1$$

$$\quad ; \qquad T_{m}(z) = (-1)^{m}cosh[m*arcosh(-z)] \qquad z<-1$$

$$$$

3. $$\theta_{m} = \displaystyle\frac{\pi}{4}\Bigg(2 - \frac{\Delta f}{f_{0}} \Bigg)$$

$$$$

4. $$\theta = \beta_{n}d_{n} = \displaystyle \Bigg(\frac{2\pi}{\lambda_{n}}\Bigg)\Bigg(\frac{\lambda_{n_{0}}}{4}\Bigg)=\displaystyle\frac{\pi}{2}\Bigg(\frac{f}{f_{0}}\Bigg)$$

$$$$

5. $$5 GHz \le f \le 30GHz$$

$$$$

6. $$Ïµ_{r_{3}} = Ïµ_{r_{2}} + 3 = 6.7$$

```python
#@title a) $f_{0} = 10$ GHz

f_min = 5 #@param
f_max = 30 #@param
n_size = 1000 #@param
f0_value = 10 #@param
e_L = 6.7 #@param
N = 2 #@param
fractional_bandwidth = 0.375 #@param



frequencies = np.linspace(f_min, f_max, n_size)
theta_m = (np.pi/4)*(2-fractional_bandwidth)
eta_L = 377/np.sqrt(e_L)
exp_1 = (eta_L-377)/(eta_L+377)

exp_2 = lambda f : np.exp(-1j*N*((np.pi*f)/(2*f0_value)))

def Tmz(m,z):

  if(z<1 and z> -1):
    return math.cos(m*math.acos(z))
  elif(z>=1):
    return math.cosh(m*math.acosh(z))
  else:
    return ((-1)**m)*math.cosh(m*math.acosh(-z))

tschebyscheff_aproximation = lambda f : exp_2(f)*exp_1*(Tmz(N,(math.cos(theta_m)**(-1))*math.cos(np.pi*(f/(2*f0_value))))/Tmz(N,math.cos(theta_m)**(-1)))

#tschebyscheff_aproximation = lambda f : 0.174*math.cos(2*np.pi*(f/(2*f0_value))) + 0.1503

# Calculate |Î_in| for each frequency and convert to float
gamma_in_values = []
for freq in frequencies:
    value = abs(tschebyscheff_aproximation(freq))
    gamma_in_values.append(value)

relative_frequencies = frequencies/f0_value

# Create the plot with Plotly
fig = px.line(x=relative_frequencies, y=gamma_in_values, labels={'x': 'Relative Frequency (f/f0)', 'y': 'Magnitude of the Input reflection coefficient |Î_in|'})
fig.update_layout(title='Magnitude of the Input Reflection Coefficient vs. Relative Frequency (f/f0)', xaxis_title='Relative Frequency (f/f0)', yaxis_title='Magnitude of the Input reflection coefficient |Î_in|')
fig.show()
```

```python
f_min = 5 #@param
f_max = 30 #@param
n_size = 1000 #@param
f0_value = 10 #@param
e_L = 5.7 #@param
N = 2 #@param
fractional_bandwidth = 0.375 #@param

frequencies = np.linspace(f_min, f_max, n_size)  # Frequencies from 5 GHz to 30 GHz

eta_L = 377/np.sqrt(e_L)
exp_1 = (eta_L-377)/(eta_L+377)

C_n = lambda n: (math.factorial(N)/(math.factorial(N-n)*math.factorial(n)))
bin_aproximation = lambda fm : 2**(-N) * exp_1 * sum([ C_n(n)*np.exp(-1j*n*np.pi * fm / f0_value) for n in range(N+1) ])

theta_m = (np.pi/4)*(2-fractional_bandwidth)
exp_1 = (eta_L-377)/(eta_L+377)
exp_2 = lambda f : np.exp(-1j*N*((np.pi*f)/(2*f0_value)))

def Tmz(m,z):

  if(z<1 and z> -1):
    return math.cos(m*math.acos(z))
  elif(z>=1):
    return math.cosh(m*math.acosh(z))
  else:
    return ((-1)**m)*math.cosh(m*math.acosh(-z))

tschebyscheff_aproximation = lambda f : exp_2(f)*exp_1*(Tmz(N,(math.cos(theta_m)**(-1))*math.cos(np.pi*(f/(2*f0_value))))/Tmz(N,math.cos(theta_m)**(-1)))


# Calculate |Î_in| for each frequency and convert to float
gamma_in_values_binomial = []
for freq in frequencies:
    value = abs(bin_aproximation(freq))
    gamma_in_values_binomial.append(value)


# Calculate |Î_in| for each frequency and convert to float
gamma_in_values_tch = []
for freq in frequencies:
    value = abs(tschebyscheff_aproximation(freq))
    gamma_in_values_tch.append(value)



relative_frequencies = frequencies/f0_value

fig = go.Figure()

y_highlight1 = 0.3
y_highlight2 = 0.4


fig.add_trace(go.Scatter(x=relative_frequencies, y=gamma_in_values_binomial, mode='lines', name='Binomial', line=dict(color='green', dash='dot')))
fig.add_trace(go.Scatter(x=relative_frequencies, y=gamma_in_values_tch, mode='lines', name='Tschebyscheff', line=dict(color='blue', dash='dashdot')))

fig.update_layout(
    title='Magnitude of the Input Reflection Coefficient vs. Relative Frequency (f/f0)',
    xaxis_title='Relative Frequency (f/f0)',
    yaxis_title='Magnitude of the Input reflection coefficient |Î_in|',
    legend=dict(x=0, y=1, traceorder='normal', orientation='h')
)

fig.show()
```
